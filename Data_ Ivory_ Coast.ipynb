```python
#Import the tools

import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn
%matplotlib inline
print("Tools Imported")
```

```python
#Import the data

data=pd.read_csv("/Users/pm3796gb/Desktop/Nova Cloud A.I/Ivory_Coast.csv")
```

```python
#Fill NaN ( Not a Number) with 0 (zero)

data=data.fillna(0)
```

```python
#Remove the 'Country Name' columns since we already know the data is from the Ivory Coast

data=data.drop("Country Name",1)
```

```python
#Take a loook at the data

data.head()
```

```python
#Set index as 'Indicator Name'
data=data.set_index("Indicator Name")
```

```python
#take a look at the 10 first rows

data.head(10)
```

```python
df=data.transpose()
df=df.reset_index()
df=df.set_index('index')
df.head()
```

# The dataset is ready to be analyse
Let's look at what happened during the past 20 years.

# Urban population vs Rural population
from 1997 to 2017 (20 years)

```python
df1=df.loc[['1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017'],['Urban population','Rural population']]
df1.head()
```

```python
df1g=df1.plot(kind='bar',figsize=(10,7),width=0.7)
df1g.set_title('year')
df1g.set_xlabel('Years')
df1g.set_ylabel('Populaion in million')
df1g.set_title('Urban population vs Rural population')
```

The graph shows a sharp increase in the urbane population since 1997.
In 2016 the rural populaions is almost the same as the urbane population.
In 2017 there are more people living in town than in the villages.
