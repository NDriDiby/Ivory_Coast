# Urban population vs Rural population
from 1997 to 2017 (20 years)


```python
df1=df.loc[['1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017'],['Urban population','Rural population']]
df1.tail()
```

```python
df1g=df1.plot(kind='bar',figsize=(10,7),width=0.7)
df1g.set_title('year')
df1g.set_xlabel('Years')
df1g.set_ylabel('Populaion in million')
```

# The graph shows a sharp increase in the urbane population since 1997.
# In 2016 the rural populaions is almost the same as the urbane population.
# In 2017 there are more living people in towns than in the villages.
