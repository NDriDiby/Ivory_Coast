# Let's put all the data on the Ivory Coast Map
# Get the latitude and longitude on google.

```python
latitude=7.5400 
longitude=5.5471
```

```python
#let's call the map
#let's add the town where the popluation is less than 100000
#we'll need a for loop
#let's add the town of the second group
#let's add the town of the third group
Ivorymap= folium.Map([latitude,longitude], zoom_start=7)

for lat,long,vil in zip(first_group.Latitude,first_group.Longitude,first_group.Ville):
    folium.Marker([lat,long], tooltip=vil).add_to(Ivorymap)

for lat,long,vil in zip(second_group.Latitude,second_group.Longitude,second_group.Ville):
    folium.Marker([lat,long], tooltip=vil,icon=folium.Icon(color="green")).add_to(Ivorymap)

for lat,long,vil in zip(third_group.Latitude,third_group.Longitude,third_group.Ville):
    folium.Marker([lat,long], tooltip=vil,icon=folium.Icon(color="red")).add_to(Ivorymap)
```
